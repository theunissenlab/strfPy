```python
import numpy as np
#from scipy.special import binom
from scipy.special import genlaguerre
#from scipy.special import hyp1f1
import matplotlib.pyplot as plt
import os

figDir = '/Users/frederictheunissen/Desktop'
```

```python
# Plot the Weighted Generalized Laguerre Polynomials
x = np.arange(0, 20.0, 0.01)

# ax.set_ylim(-1.0, 1.0)
alpha = 4
order = 5
plt.figure(figsize=(3,2), dpi = 100)
plt.title(f'Laguerre alpha={alpha}')
for xorder in np.arange(0, order):
    y = np.exp(-x)*np.power(x, alpha)*genlaguerre(xorder, alpha)(x)
    ynorm = np.sqrt(np.sum(y*y))
    y = y/ynorm
    plt.plot(x, y, label=f'{xorder}')
# plt.legend(loc='best')
plt.savefig(os.path.join(figDir, 'Laguerre.eps'))
```

```python
# Plot the Lanczos Smoothing Filter
x = np.arange(-10.0, 10.0, 0.01)

alpha = 20   # This is the size of the window
tau = 2  # This is the period of the internal sync
plt.figure(figsize=(3,2), dpi = 100)
plt.title(f'Lanczos Tau={tau}')
y = alpha*tau*np.sin(np.pi*x/tau)*np.sin(np.pi*x/(alpha))/(np.pi**2*x**2)

plt.plot(x, y, 'k', label=f'{xorder}')
# plt.legend(loc='best')
plt.savefig(os.path.join(figDir, 'Lanczos.eps'))
```
