```python
# General dependencies
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
import pickle as pkl
```

```python
# Data folder
# Change this to your local path where the ECoG data is stored.

data_foler = '/Users/frederictheunissen/Google Drive/My Drive/PowerData/ECoG'
data_files = ['ECOG_spec_fband10.pkl', 'ECOG_spec_fband25.pkl', 'ECOG_spec_fband50.pkl']
```

```python
## Load the first data set - you will want to do the calculation for all 3 data sets. 

with open(f'{data_foler}/{data_files[0]}', 'rb') as f:
    ECOG_data = pkl.load(f)
```

```python
# Plot the waveforms for each band and trial

ntrials = len(ECOG_data)
nbands = ECOG_data[0].shape[0]
ntimes = ECOG_data[0].shape[1]

fig, axes = plt.subplots(nbands, 1, figsize=(10, 5*nbands))
for ib in range(nbands):
    for i in range(ntrials):
        axes[ib].plot(ECOG_data[i][ib, :], alpha=0.5)
    axes[ib].set_title(f'Band {ib+1}')
    axes[ib].set_xlabel('Time (samples)')
    axes[ib].set_ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
